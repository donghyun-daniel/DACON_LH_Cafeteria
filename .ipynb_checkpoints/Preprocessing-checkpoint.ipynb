```python
import numpy as np
import pandas as pd
```

```python
data = {}
for year in range(2016, 2022):
    data[year] = pd.read_csv("data/기상정보/" + str(year) + "_진주.csv", encoding = "cp949")
```

```python
data[2016].head()
```

```python
data[2021].head()
```

```python
for year in range(2016, 2022):
    data[year] = data[year][["일시", "최저기온(°C)", "평균기온(°C)","최고기온(°C)",
                             "평균 풍속(m/s)", "최대 풍속(m/s)", "일강수량(mm)"]]
```

```python
for year in range(2016, 2022):
    data[year] = data[year].fillna(0)
```

```python
data_concat = pd.concat(data).reset_index(drop = True)
```

```python
data_concat.to_csv('data/진주기상정보.csv', index=False, encoding='cp949')
```
